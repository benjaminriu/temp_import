# Check results

```python
task_name = "regression" #else classification
metrics = ["R2", "R2train"] if task_name == "regression" else ["ACC", "AUC", "ACCtrain", "AUCtrain"]
```

```python
import pandas as pd
result_df = pd.read_csv(output_repository+task_name+benchmark_output_file, index_col = 0)
```

```python
result_df = result_df[result_df["success"]==True]
for metric in metrics:
    result_df[metric] = result_df[metric].astype(float)
```

```python
#best of all
result_df.groupby(["dataset","method"]).mean().sort_values(["dataset", metrics[0]]).reset_index().groupby("dataset").last()
```

```python
#best of one
dataset_index = 0
result_df.groupby(["dataset","method"]).mean().sort_values(["dataset", metrics[0]]).reset_index().set_index("dataset").loc[dataset_index]
```
